```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv('clean_output.csv')
data
```

```python
data.boxplot(figsize=(8,6))
plt.title('Boxplots of all features')
plt.xlabel('Feature')
plt.show()
```

```python
# Create histograms for all features
plt.figure(figsize=(10, 6))

for col in data.columns:
    plt.hist(data[col], bins=10, alpha=0.5, label=col)
    

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histograms of All Features")
plt.xscale('log')
plt.legend()
plt.grid()
plt.show()
```

```python
import seaborn as sns
# Create density plots for all features
plt.figure(figsize=(10, 6))

for col in data.columns:
    sns.kdeplot(data[col], label=col, fill=True, alpha=0.5)

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Density Plots of All Features")
plt.xscale('log')
plt.legend()
plt.grid()
plt.show()
```
